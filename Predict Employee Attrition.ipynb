## Label Encoding
In machine learning, we ususally deal with datasets that contain multiple labels in one or more than one columns. These labels can be in the form of words or numbers. To make the data understandable or in human-readable form, the training data is usually labelled in words.

```python
# Target Variable(Attrition)
data['Attrition'] =  data['Attrition'].replace({'No' : 0, 'Yes' : 1})
```

```python
# encode binary variables
data['OverTime'] = data['OverTime'].map({'No' : 0, 'Yes' : 1})
data['Gender'] = data['Gender'].map({'Male' : 0, 'Female' : 1})
```

```python
data['Over18'] = data['Over18'].map({'Y' : 1, 'No' : 0})
```

```python
# encode categorical columns which are ordinal, use label encoding
# apply label encoder to df_categorical
from sklearn.preprocessing import LabelEncoder
encoding_cols = ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']
label_encoders = {}
for column in encoding_cols:
    label_encoders[column] = LabelEncoder()
    data[column] = label_encoders[column].fit_transform(data[column])
```

# Training and Testing Spliting Data

```python
X = data.drop(['Attrition'], axis=1)
y = data['Attrition'].values
```

## Resampling
This is a method that consists of drawing repeated samples from the original data samples. The method of resampling is a nonparametic method of statistical inference. Oversampling and undersampling in data analysis are techniques used to adjust the class distribution of a dataset.

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print(Counter(y))
rus = RandomOverSampler(random_state = 42)
X_over, y_over = rus.fit_resample(X,y)
print(Counter(y_over))
```

```python
# Split the data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_over, y_over, test_size=0.2, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
xtrain_scaled = std_scaler.fit_transform(X_train)
xtest_scaled = std_scaler.transform(X_test)
```

```python
xtrain_scaled
xtest_scaled
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
Random_forest_model = RandomForestClassifier()
```

```python
Random_forest_model.fit(xtrain_scaled, y_train)
```

```python
y_pred = Random_forest_model.predict(xtest_scaled)
```

## Model Evaluation

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
sns.heatmap(cm, annot=True, fmt='d')
```

```python
# Accuracy
from sklearn.metrics import accuracy_score
model_accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of the model = ", model_accuracy)
```

```python
# Precision
from sklearn.metrics import precision_score
model_precision = precision_score(y_test, y_pred)
print("Precision score of the model = ", model_precision)
```

```python
# Recall
from sklearn.metrics import recall_score
model_recall = recall_score(y_test, y_pred)
print("Recall score of the model = ", model_recall)
```

```python
# F1-Score
from sklearn.metrics import f1_score
f1score = f1_score(y_test, y_pred)
print("F1 score of the model = ", f1score)
```

```python
# Classification Report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
prediction = logreg.predict(X_test)
cnf_matrix = confusion_matrix(y_test, prediction)
print('Accuracy Score :', accuracy_score(y_test, prediction))
```

```python
fig = plt.figure(figsize = (15, 6))
ax1 = fig.add_subplot(1,2,1)
ax1 = sns.heatmap(pd.DataFrame(cnf_matrix), annot = True, cmap = 'Blues', fmt = 'd')
bottom, top = ax1.get_ylim()
ax1.set_ylim(bottom + 0.5, top - 0.5)
plt.xlabel('Predicted')
plt.ylabel('Expected')

ax2 = fig.add_subplot(1,2,2)
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = roc_curve(y_test, prediction)
auc = roc_auc_score(y_test, prediction)
ax2 = plt.plot(fpr,tpr,label='data 1, auc='+str(auc))
plt.legend(loc=4)
plt.show()
```
